```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
import urllib.request
import os.path
import zipfile
data_url = "https://www.ssa.gov/oact/babynames/state/namesbystate.zip"
local_filename = "babynamesbystate.zip"
if not os.path.exists(local_filename):
  with urllib.request.urlopen(data_url) as resp, open(local_filename,'wb') as f:
    f.write(resp.read())
zf = zipfile.ZipFile(local_filename, 'r')
ca_name = 'STATE.CA.TXT'
field_names = ['State', 'Sex', 'Year', 'Name', 'Count']
with zf.open(ca_name) as fh:
  babynames = pd.read_csv(fh, header=None, names=field_names)
babynames.head()
```

```python
babynames_length=babynames["Name"].str.len()
babynames["Name_Length"]= babynames_length
babynames
```

```python
babynames.sort_values("Name_Length", ascending=True)
babynames
```

```python
babynames=babynames.drop(["Name_Length"], axis=1)
babynames
```

```python
def dr_ea_count(string):
  return string.count("dr")+string.count("ea")

babynames['dr_ea_count'] = babynames['Name'].map(dr_ea_count)
babynames=babynames.sort_values(by="dr_ea_count", ascending=False)
babynames
```

```python
babynames=babynames.drop(["dr_ea_count"], axis=1)
babynames
```

```python
female_babynames = babynames[babynames['Sex'] == 'F']
female_babynames
```

```python
fig=px.line(female_babynames[female_babynames["Name"]=="Jennifer"], x="Year", y= "Count")
fig.update_layout(font_size=18)
fig
```

```python
fig=px.line(female_babynames[female_babynames["Name"]=="Jennifer"].sort_values(["Year","Count"]), x="Year", y= "Count")
fig.update_layout(font_size=18)
fig
```
